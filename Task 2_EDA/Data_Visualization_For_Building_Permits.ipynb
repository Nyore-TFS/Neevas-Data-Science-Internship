```python
data['Valuation($)'] = data['Valuation($)'].astype(int)
data.info()
```

#### Determine the total valution gotten by the top contracted contractors.

```python
top_contractors = data["Contractor's Full Name"].value_counts().nlargest(13)
top_contractors_data = df[data["Contractor's Full Name"].isin(
    top_contractors.index)]
valuation_sum = top_contractors_data.groupby(
    "Contractor's Full Name")["Valuation($)"].sum().reset_index()
valuation_sum = valuation_sum.sort_values("Valuation($)",
                                          ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x="Contractor's Full Name", y="Valuation($)",
            data = valuation_sum)
plt.title("Top Contractor's by Frequeuncy and Project Valuaiton")
plt.xlabel("Contractor's Name")
plt.ylabel('Valuation($)')
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.ticklabel_format(axis='y', style='plain', useOffset=False)
plt.show()
```

#### Determine the total amount paid to contractors monthly form August 2022 to August 2024 which is the duration covered by the dataset.

```python
valuation_per_month = data.groupby(
    "Month")["Valuation($)"].sum().reset_index()
valuation_per_month = valuation_per_month.sort_values("Valuation($)", 
                                                      ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x="Month", y="Valuation($)",
            data = valuation_per_month)
plt.title("Project Valuation per Month")
plt.xlabel("Month's")
plt.ylabel('Valuation($)')
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.ticklabel_format(axis='y', style='plain', useOffset=False)
plt.show()
```

#### Determine the total amount spent on project for the years captured in dataset

```python
# Extract year from the date
data['Year'] = pd.to_datetime(data['Date Issued']).dt.year

# Group by year and sum valuations
yearly_valuation = data.groupby('Year')['Valuation($)'].sum().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 8))
sns.barplot(x='Year', y='Valuation($)', data=yearly_valuation)

# Add titles and labels
plt.title('Yearly Project Valuation (2022-2024)')
plt.xlabel('Year')
plt.ylabel('Total Valuation')
plt.yticks(rotation=45)
plt.ticklabel_format(axis='y', style='plain', useOffset=False)
plt.tight_layout()

# Show the plot
plt.show()
```

#### Viaualization for year 2023 to see which quarter of the year in which more money was spent on contruction works.

```python
data['Year_2023'] = pd.to_datetime(data['Date Issued'])

# Filter for the year 2023
data_2023 = data[data['Year_2023'].dt.year == 2023]

# Create a new column for quarters
data_2023['Quarter'] = data_2023['Year_2023'].apply(
    lambda x: 'Jan-Jun' if x.month <= 6 else 'Jul-Dec'
    )

# Group by quarters and sum valuations
quarterly_valuation = data_2023.groupby('Quarter')['Valuation($)'].sum().reset_index()


plt.figure(figsize=(8, 6))
sns.barplot(x='Quarter', y='Valuation($)', data=quarterly_valuation)


plt.title('2023 Project Valuation by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Valuation')
plt.yticks(rotation=45)
plt.ticklabel_format(axis='y', style='plain', useOffset=False)
plt.tight_layout()
```

#### Visaualizing the total valuation based on permit type

```python
permit = data.groupby(
    "Permit Type")["Valuation($)"].sum().reset_index()

sns.lineplot(x='Permit Type', y='Valuation($)', data=permit)
plt.xlabel('Permit Type')
plt.xticks(rotation=90)
plt.ylabel('Valuation($)')
plt.title('Line Chart of Valuation by Permit Type')
plt.ticklabel_format(axis='y', style='plain', useOffset=False)
plt.show()
```
